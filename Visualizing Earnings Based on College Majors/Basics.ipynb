# Visualizing Earnings Based on College Majors
In this project I am working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the data and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors). <br>
<br>
Each row in the dataset represents a dufferent major in college and contains information on gender diversity, employment rates, median salaries, and more.
<br>
<br>
Using visualization with pandas plotting functionality and Jupyter notebook interface will allow me to explore questions from the data set such as:
 - Do students in more popular majors make more money?
 - How many majors are predominantly male/female?
 - Which category of majors have the most students?

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads=pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
raw_data_count=5
recent_grads=recent_grads.dropna()
print(recent_grads.head())
```

# Using Scatter Plots to Explore Data
Here, I will be using scatter plots to explore the following questions:
 - Do students in more popular majors make more money?
 - Do students that majored in subjects that were majority female make more money?
 - Is there a link between the number of full-time employees and median salary?

```python
ax=recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median Salary vs Sample Size')
ax.set_xlabel('Sample Size')
ax.set_ylabel('Median Salary ($)')
```

```python
ax=recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs Sample Size')
ax.set_xlabel('Sample Size')
ax.set_ylabel('Unemployment Rate')
```

```python
ax=recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full Time Employment vs Median Salary')
ax.set_xlabel('Full Time Eployees')
ax.set_ylabel('Median Salary($)')
```

```python
ax=recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Females per Major vs Unemployment Rate')
ax.set_xlabel('Feamles per Major')
ax.set_ylabel('Unemployment Rate')
```

```python
ax=recent_grads.plot(x='Men', y='Median', kind='scatter', title='Male Graduates vs Median Salary')
ax.set_xlabel('Male Graduates')
ax.set_ylabel('Median Salary ($)')
```

```python
ax=recent_grads.plot(x='Women', y='Median', kind='scatter', title='Female Graduates vs Median Salary')
ax.set_xlabel('Female Graduates')
ax.set_ylabel('Median Salary ($)')
```

Now that the scatter plots are all layed out, it is easy to analyze the data from these visualizations. From the scatter plots, we can conclude that male graduates earn more than female graduates.  Female graduates range from 1 to 350,000, whereas male graduates ranged from 1 to 200,000. Thus, there is more information based on female graduates. In addition, females per major have weak correlations with the unemployment rate, whereas the entire sample size has a stronger correlation with the unemployment rate. <br>
<br>
The scatter plots unfortunately only give us a cursory understanding of the distributions of the average salary of female and male graduates.

# Using Histograms to Explore Data
Here, I explore the distribution of values in a column using histograms. Here I use these plots to explore the following questions:
 - What percent of majors are predominantly male/female?
 - What's the most common median salary?

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
titles=["Sample Size", "Median Salaries", "Employed", "Employed Full Time", "Women per Major", "Unemployment Rate", "Male Graduates", "Female Graduates"]
fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, title=titles[r])
    ax= fig.tight_layout()
    
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
titles=["Sample Size", "Median Salaries", "Employed", "Employed Full Time", "Women per Major", "Unemployment Rate", "Male Graduates", "Female Graduates"]
fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, title=titles[r])
    ax= fig.tight_layout()
```

From these box plots you can see the most common median is between \$30,000 and \$40,000 a year with the majority of graduates employed full time. It appears that 93% of majors are predominantly male while 97% is predominantly female.

# Using Scatter Matrix Plots to Explore Data
A scatter matrix plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously. A scatter matrix plot consists of n by n plots on a grid, where n is the number of columns, the plots on the diagonal are histograms, and the non-diagonal plots are scatter plots. We will explore the questions from the previous graphs to see if we can conclude to more cohesive answers.

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6, 6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Full_time']], figsize=(10, 10))
```

We can see that there is a strong and positive correlation of full time employees assocciated with the sample size, however there is a weak correlation of full time employees to the median salary. Let's see if there is a stronger correlation among genders.

```python
scatter_matrix(recent_grads[['Women', 'Men', 'Median']], figsize=(10,10))
```

It appears that there is a weak correlation among the median salaries, however there is a stronger correlation for men to have a hogher median salary.

```python
scatter_matrix(recent_grads[['Women', 'Men', 'Unemployment_rate']], figsize=(10,10))
```

Here, we can see men more frequently are affected by the unemployment rate as opposed to women. It should also be noted that more women are involved in this dataset than men.
